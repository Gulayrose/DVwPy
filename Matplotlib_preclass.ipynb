```python
import numpy as np   
import matplotlib.pyplot as plt                           
import random 
%matplotlib inline  
```

```python
# Functional Method
x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

plt.subplot(2,2,1)
plt.plot(x,y1, 'ro');
plt.title('$y_1=x$');

plt.subplot(2,2,2)
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$');

plt.subplot(2,2,3)
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(2,2,4)
plt.plot(x,y4, 'ks');
plt.title('$y_4=\sqrt{x}$');
```

```python
x = np.linspace(0,10,10) 
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sqrt(x)  # karekök

# plt.figure(figsize=(14,10))
plt.subplot(4,1,1)
plt.plot(x,y1, 'ro');
plt.title('$y_1=x$');

plt.subplot(4,1,2)
plt.plot(x,y2, 'g--');
plt.title('$y_2=x^2$');

plt.subplot(4,1,3)
plt.plot(x,y3,'b^');
plt.title('$y_3=x^3$');

plt.subplot(4,1,4)
plt.plot(x,y4, 'ks');
plt.title('$y_4=\sqrt{x}$');
```

```python
company=['Googl', 'Amzn', 'Msft', 'Fb']
revenue=[90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(company,revenue);
```

```python
plt.bar(xpos,revenue);
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("Tech Comp")
plt.bar(xpos,revenue, label="revenue");
plt.bar(xpos, profit,label="profit")
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("Tech Comp")
plt.bar(xpos-0.2,revenue, label="revenue");
plt.bar(xpos+0.2, profit,label="profit")
plt.legend()
```

```python
plt.yticks(xpos, company)
plt.ylabel("revenue(bln)")
plt.title("Tech Comp")
plt.barh(xpos-0.2,revenue, label="revenue");
plt.barh(xpos+0.2, profit,label="profit")
plt.legend()
```

```python
## Histogram
# matplotlib.pyplot.hist(x, bins=None, range=None, density=False, 
# orientation='vertical', color=None, label=None, stacked=False,data=None)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar);
```

```python
plt.hist(blood_sugar, bins=3, rwidth=0.9);
```

```python
plt.hist(blood_sugar, bins=5, rwidth=0.9);
```

```python
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.9, color='g', histtype='step');
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.9, color=['green', 'orange']),
label=['men', 'women']
plt.legend()
```

```python
## Pie Chart
# matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, 
# pctdistance=0.6, labeldistance=1.1, startangle=0, radius=1, data=None)
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.1f%%', shadow=True, explode=[0,0,0.5,0,0])
plt.show()
```
